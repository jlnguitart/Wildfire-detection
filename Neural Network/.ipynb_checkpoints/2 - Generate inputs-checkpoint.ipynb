```python

class Hexint:
    
    
    def __init__(self, value=0, base7=True):
        #error is less than zero
        if value<0:
            raise 'BadHexintError'
        if base7==True:
            #any non 0-6 digit will raise error
            temp = '%ld'%value
            for i in range(len(temp)):
                if int(temp[i])>6:
                    raise 'BadHexintError'
            self.numval=value
        else:
            temp = 0
            mul = 1
            while value>0:
                temp += (value%7)*mul
                value /= 7
                mul *= 10
            self.numval=temp
        self.numstr = '%ld'%self.numval
        self.ndigits = len( self.numstr)
    # convert to a string for displaying result
    def __str__ (self):
        return '('+self.numstr+')'
        
    __repr__ = __str__


    def __hash__(self):
        return hash((self.numstr, self.numval))    
    #comparaison ==
    
    def __eq__ (self, object):
        if isinstance( object, Hexint):
            return self.numval==object.numval
        else:
            return False
            
    # Hexint addition
    
    def __add__ (self,object):
        #lookup table
        A = [ [0,  1,  2,  3,  4,  5,  6],
              [1, 63, 15,  2,  0,  6, 64],
              [2, 15, 14, 26,  3,  0,  1],
              [3,  2, 26, 25, 31,  4,  0],
              [4,  0,  3, 31, 36, 42,  5],
              [5,  6,  0,  4, 42, 41, 53],
              [6, 64,  1,  0,  5, 53, 52] ]
        #pad out with zeros to make strs same length
        slen = self.ndigits-object.ndigits
        if slen>0:
            numa = '0'+self.numstr
            numb = '0'+('0'*slen)+object.numstr
        else:
            numa = '0'+('0'*(-slen))+self.numstr
            numb = '0'+object.numstr
        maxlen = len(numa)
        
        total = 0
        mul = 1
        for i in range (maxlen):
            ii = maxlen-i-1
            t = A[ int(numa[ii])][int(numb[ii])]
            total += (t%10)*mul
            carry = int(t/10)
            
            for j in range(i+1,maxlen):
                jj=maxlen-j-1
                if carry>0:
                    t = A[ int(numa[jj])][carry]
                    numa = numa[:jj]+str(t%10)+numa[jj+1:]
                    carry = int(t/10)
            mul *= 10
            
        return Hexint( total )
            
    def __mul__ ( self, object):
        if isinstance(object,Hexint):
            #lookup table
            M = [ [ 0, 0, 0, 0, 0, 0, 0],
                  [ 0, 1, 2, 3, 4, 5, 6],
                  [ 0, 2, 3, 4, 5, 6, 1] , 
                  [ 0, 3, 4, 5, 6, 1, 2] , 
                  [ 0, 4, 5, 6, 1, 2, 3] , 
                  [ 0, 5, 6, 1, 2, 3, 4] , 
                  [ 0, 6, 1, 2, 3, 4, 5] ]
            #ZERO PADDING
            slen = self.ndigits-object.ndigits
            if slen>0:
                numa = '0'+self.numstr
                numb = '0'+('0'*slen)+object.numstr
            else:
                numa = '0'+('0'*(-slen))+self.numstr
                numb = '0'+object.numstr
                maxlen = len(numa)    
                powers = [10**i for i in range(maxlen)]
                  
            sum = Hexint(0)
        
            for i in range(maxlen):
                ii = maxlen-i-1
                partial = int(0)
                mul = powers[i]
                for j in range(maxlen):
                    jj = maxlen-j-1
                    if numa[ii]!=0:
                        partial += M[ int(numa[ii])][int(numb[jj])]*mul
                    mul *= 10
                sum += Hexint(partial)
            return sum
        
        # scalar multiplication
        elif isinstance(object,int):
            if object>0:
                num = Hexint(self.numval)
            else:
                num = -Hexint(self.numval)
                
            total = Hexint(0)
            for i in range(abs(object)):
                total += num
        return total
    __rmul=__mul__
    
    #negate a Hexint
    
    def __neg__( self ): 
        total = 0
        mul = 1
        for i in range(self.ndigits -1,-1,-1):
            if self.numstr[i]=='1': 
                total += (4*mul) 
            elif self.numstr[i]=='2': 
                total += (5*mul) 
            elif self.numstr[i]=='3': 
                total += (6*mul) 
            elif self.numstr[i]=='4': 
                total += (1*mul) 
            elif self.numstr[i]=='5':
                total += (2*mul) 
            elif self.numstr[i]=='6': 
                total += (3*mul) 
            mul *= 10
        return Hexint( total )
   
    # Hexint subtraction 
    def __sub__( self, object ): 
        return self + (-object)   
    
        
    # get the digit of the Hexint at position pos 
    def __getitem__( self, pos ): 
        if pos>=self.ndigits: 
            raise 'HexIndexError' 
        else: return int( self.numstr[self.ndigits -pos -1] )
    
    # get a Hexint that is some part of the original 
    def __getslice__( self, low,high ): 
        return Hexint( int(self.numstr[low:high]) )
        
    # number of digits in the Hexint 
    def __len__( self ): 
        return self.ndigits
        
    # return spatial integer coord pair 
    def getSpatial( self ): 
        xi,yi,zi = self.getHer( ) 
        # return result 
        return ( (xi + yi - 2*zi)/3, (-xi + 2*yi - zi )/3 ) 
        
        
    # returns a 3-tuple using Herâ€™s coord system 
    def getHer( self ): 
        xc,yc,zc = 1,0,-1 
        x,y,z = 0,0,0
        for i in range(self.ndigits -1,-1,-1): 
            if i<self.ndigits -1: 
                # compute key points 
                xc,yc,zc = (4*xc - 5*yc + zc)/3, (xc + 4*yc - 5*zc)/3, (-5*xc + yc + 4*zc)/3 
            #compute the rotation
            if self.numstr[i]=='1':
                x += xc 
                y += yc 
                z += zc 
            elif self.numstr[i]=='2':
                x -= yc 
                y -= zc 
                z -= xc 
            elif self.numstr[i]=='3': 
                x += zc 
                y += xc 
                z += yc 
            elif self.numstr[i]=='4': 
                x -= xc 
                y -= yc 
                z -= zc 
            elif self.numstr[i]=='5': 
                x += yc 
                y += zc 
                z += xc 
            elif self.numstr[i]=='6': 
                x -= zc 
                y -= xc 
                z -= yc  
        #return result
        return (x,y,z)
```

```python
lkup_dir='/home/jnueno/Final_Project/Model/Lookups/'

fire_lkup = pd.read_excel(lkup_dir+"fire_file_lkup.xlsx")

shp_path='/home/jnueno/Final_Project/HexGrid US/Fires/Shp/'

california_fires=fire_lkup[fire_lkup['final_id'].isin(allFiles)]
```

```python

idx = california_fires.groupby(['fire_id'])['observation date'].transform(max) == california_fires['observation date']
last_day=california_fires[idx]
idx2 = last_day.groupby(['fire_id'])['observation time'].transform(max) == last_day['observation time']
final_perimeter=last_day[idx2]

idx = california_fires.groupby(['fire_id'])['observation date'].transform(min) == california_fires['observation date']
first_day=california_fires[idx]
idx2 = first_day.groupby(['fire_id'])['observation time'].transform(min) == first_day['observation time']
initial_perimeter=first_day[idx2]

initial_perimeter_merge=initial_perimeter[['fire_id','observation date', 'observation time']]
initial_perimeter_merge.columns=['fire_id','initial date', 'initial time']

california_fires = california_fires.merge(initial_perimeter_merge, how='left', on='fire_id')


california_fires['timestamp']=(pd.to_datetime(california_fires['observation date'],infer_datetime_format=True)-pd.to_datetime(california_fires['initial date'],infer_datetime_format=True)).dt.days*24+(california_fires['observation time'].apply(int)-california_fires['initial time'].apply(int))

fire_observations=california_fires.groupby(['fire_id'])['observation date'].count().reset_index()
multiple_observations=fire_observations[fire_observations['observation date']>1]
#multiple_observations=fire_observations[fire_observations['observation date']==23]

target_fires=california_fires[california_fires['fire_id'].isin(multiple_observations['fire_id'].values)]




#california_fires['timestamp']=*24+(california_fires['observation time'].apply(int)-california_fires['initial time'].apply(int))
```

```python


for i,fire in multiple_observations.iterrows():
    
    fire_index=target_fires[target_fires['fire_id']==fire['fire_id']]
    
    final_perimeter_target=final_perimeter[final_perimeter['fire_id']==fire['fire_id']]['final_id']
    final_perimeter_timestamp=fire_index[fire_index['final_id']==final_perimeter_target.iloc[0]]['timestamp']
    
    
    fire_master=pd.read_csv(flat_data_dir+str(final_perimeter_target.iloc[0])+'.csv')
    
    fire_master['fire_start']=999999
    fire_master.loc[fire_master['on_fire']==1,'fire_start'] = final_perimeter_timestamp.iloc[0]
    
    
    for k,perimeter in fire_index.iterrows():
        
        #print(perimeter)
        
        if perimeter['final_id']!=final_perimeter_target.iloc[0]:
            
            perimeter_data=pd.read_csv(flat_data_dir+str(perimeter['final_id'])+'.csv')
            
            perimeter_data=perimeter_data[perimeter_data['on_fire']==1]

            perimeter_data['timestamp_tmp']=perimeter['timestamp']
            perimeter_data=perimeter_data[['us_gid_to_','timestamp_tmp']]
            fire_master=fire_master.merge(perimeter_data, how='left', on='us_gid_to_')
            fire_master['fire_start']=fire_master[['fire_start','timestamp_tmp']].min(axis=1)

            fire_master.drop(['timestamp_tmp'], axis=1, inplace=True)
            
    fire_master.to_csv(input_data_dir+fire['fire_id']+'.csv')

            
    

```

```python
def get_boundary(gbt_idx):
    boundary=[]
    gbt=gbt_idx
    gbt_hi=gbt.apply(Hexint)
    
    for hi in gbt_hi:
        b=0
        for i in range(1,7):
            if all(gbt_hi!=hi+Hexint(i)):
                boundary.append(True)
                b=1
                break
        if b==0:
            boundary.append(False)

    #print(boundary)
    return gbt_idx[boundary]

def calculate_ring(dest,origin):
    
    h1=Hexint(dest)
    h2=Hexint(origin)
    
    distHer=(h1-h2).getHer()
    
    return max(abs(max(distHer)),abs(min(distHer)))

def import_initial_fire_shapefile(fire_sample):
    
    geodf=gpd.read_file(shp_path+fire_sample+'_hex_polygon.shp')
    cols=geodf.columns
    flatdf=pd.DataFrame(geodf.values)
    flatdf.columns=cols
    
    return flatdf
    
def generate_rings(fire):
    
    initial_perimeter_target=initial_perimeter[initial_perimeter['fire_id']==fire]['final_id']
    perimeter_0=import_initial_fire_shapefile(str(initial_perimeter_target.iloc[0]))
    fire_master=pd.read_csv(input_data_dir+fire+'.csv')
    #print(fire_master.columns)
    boundary=get_boundary(perimeter_0['gbt_addr'])
    df = pd.DataFrame(columns=['us_gid_to_','ring'])
    
    for i,row in fire_master.iterrows():
    
        ring=999999
        dest=row['us_gid_to_']
        
        for k,origin in boundary.iteritems():
            if ring>calculate_ring(dest,origin):
                
                ring=calculate_ring(dest,origin)
        
        
        df=df.append(pd.DataFrame({'us_gid_to_': [dest], 'ring': [ring]}), ignore_index=True)

    fire_master.drop(columns=['ring'], inplace=True)
    df['us_gid_to_']=df['us_gid_to_'].apply(int)
    fire_master=fire_master.merge(df, how='left', on='us_gid_to_')
    fire_master.loc[fire_master['fire_start']==0,'ring'] = 0
    fire_master.to_csv(input_data_dir+fire+'.csv')

#generate_rings('CASRF-000914')

   
```

```python
#for i,fire in multiple_observations.iterrows():
    
    
import warnings
warnings.filterwarnings('ignore')
import multiprocessing

grid_fire_dir='/home/jupyter/data/perimeters_nn/'
flat_data_dir='/home/jupyter/data/csv_nn/'

def func(d):
    for i,row in d.iterrows():
        try:
            generate_rings(row['fire_id'])
        except:
            pass

def parallelize_dataframe(df, func):
    num_cores = multiprocessing.cpu_count()-1 
    print(num_cores)
    num_partitions = num_cores 
    df_split = np.array_split(df.sample(frac=1).reset_index(drop=True), num_partitions)
    pool = multiprocessing.Pool(num_cores)
    pool.map(func, df_split)
    pool.close()

    return df

parallelize_dataframe(multiple_observations,func)
    
```
